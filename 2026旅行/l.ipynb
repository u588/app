```python
import pandas as pd
import plotly.express as px

# 1. 读取与清洗数据
df = pd.read_excel('/home/ts/app/2026旅行/pro.xlsx')
# 确保列名无多余空格
df.columns = [c.strip() for c in df.columns]
df['金额'] = pd.to_numeric(df['金额'], errors='coerce')
df = df.dropna(subset=['金额'])

# 2. 按层级聚合并计算占比
df_grouped = df.groupby(['项目', '项目明细'])['金额'].sum().reset_index()
# 计算占上级类别百分比
df_grouped['category_total'] = df_grouped.groupby('项目')['金额'].transform('sum')
df_grouped['占上级%'] = (df_grouped['金额'] / df_grouped['category_total'] * 100).round(1)
# 计算占总金额百分比
grand_total = df_grouped['金额'].sum()
df_grouped['占总%'] = (df_grouped['金额'] / grand_total * 100).round(1)

# 3. 拼接扇形内部显示文本
df_grouped['display_text'] = df_grouped.apply(
    # lambda x: f"{x['项目明细']}<br>¥{x['金额']:,.0f}<br>{x['占上级%']}%", axis=1
    lambda x: f"{x['金额']:,.0f}<br>{x['占上级%']}%", axis=1
)

# 4. 绘制旭日图 (注意：此处已移除 text 参数)
fig = px.sunburst(
    df_grouped,
    path=['项目', '项目明细'],
    values='金额',
    title='消费支出层级占比旭日图',
    color='项目',
    hover_data={'金额': ':,.2f', '占总%': ':.2f%', '占上级%': ':.2f%'}
)

# 5. 关键修正：通过 update_traces 注入自定义文本与悬停模板
fig.update_traces(
    text=df_grouped['display_text'],       # 扇区内部文字
    # textposition='inside',                 # 文字位置
    # insidetextorientation='radial',        # 文字沿径向排列
    hovertemplate=(
        "<b>%{label}</b><br>"
        "💰 金额: ¥%{value:,.2f}<br>"
        "📊 占总金额: %{customdata[1]}<br>"
        "📁 占上级类别: %{customdata[2]}<extra></extra>"
    ),
    textfont=dict(size=11, family="Microsoft YaHei, SimHei, sans-serif"),
    # mininscribedangle=25  # 面积过小(<25°)的扇形自动隐藏文字，防重叠
)

# 6. 全局布局优化
fig.update_layout(
    title_x=0.5,
    font=dict(family="Microsoft YaHei, SimHei, sans-serif", size=12),
    width=600,
    height=600,
    margin=dict(t=70, l=10, r=10, b=10),
    sunburstcolorway=['#636efa', '#EF553B', '#00cc96', '#ab63fa'] # 可选：自定义大类配色
)

# 7. 显示图表
fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# 1. 读取与清洗
df = pd.read_excel('/home/ts/app/2026旅行/pro.xlsx')
df.columns = [c.strip() for c in df.columns]
df['金额'] = pd.to_numeric(df['金额'], errors='coerce')
df = df.dropna(subset=['金额'])

# 2. 聚合与占比计算
df_cat = df.groupby('项目')['金额'].sum().reset_index()
df_sub = df.groupby(['项目', '项目明细'])['金额'].sum().reset_index()

grand_total = df_cat['金额'].sum()
df_cat['占总%'] = df_cat['金额'] / grand_total

df_sub = df_sub.merge(df_cat[['项目', '金额']], on='项目', suffixes=('', '_cat'))
df_sub['占总%'] = df_sub['金额'] / grand_total
df_sub['占上级%'] = df_sub['金额'] / df_sub['金额_cat']

# 3. 构建底层数据 (核心修复：使用唯一 ids 避免标签冲突)
labels, parents, values, texts, ids = [], [], [], [], []
cd_total, cd_parent = [], []

# ① 根节点
ids.append('root'); parents.append('')
labels.append('总支出'); values.append(grand_total)
texts.append(f'总支出<br>¥{grand_total:,.0f}')
cd_total.append(1.0); cd_parent.append(1.0)

# ② 大类节点
for _, row in df_cat.iterrows():
    cat_id = f"cat_{row['项目']}"
    ids.append(cat_id); parents.append('root')  # 明确指向 root 的 id
    labels.append(row['项目']); values.append(row['金额'])
    texts.append(f"{row['项目']}<br>¥{row['金额']:,.0f}<br>{row['占总%']:.1%}")
    cd_total.append(row['占总%']); cd_parent.append(row['占总%'])

# ③ 明细节点
for _, row in df_sub.iterrows():
    sub_id = f"sub_{row['项目']}_{row['项目明细']}"
    cat_id = f"cat_{row['项目']}"
    ids.append(sub_id); parents.append(cat_id)  # 明确指向对应大类的 id
    labels.append(row['项目明细']); values.append(row['金额'])
    texts.append(f"{row['项目明细']}<br>¥{row['金额']:,.0f}<br>{row['占上级%']:.1%}")
    cd_total.append(row['占总%']); cd_parent.append(row['占上级%'])

# 4. 绘制旭日图 (修复白图)
fig = go.Figure(go.Sunburst(
    ids=ids,
    labels=labels,
    parents=parents,
    values=values,
    branchvalues='total',       # 父节点值 = 子节点值之和，确保角度计算正确
    text=texts,                 # 扇区内常驻文本
    textfont=dict(size=10, family="Microsoft YaHei, SimHei"),
    # mininscribedangle=20,       # 自动隐藏过小扇形(<20°)的文字，防重叠?
    customdata=np.column_stack((cd_total, cd_parent)),
    hovertemplate=(
        "<b>%{label}</b><br>"
        "💰 金额: ¥%{value:,.2f}<br>"
        "📊 占总金额: %{customdata[0]:.2%}<br>"
        "📁 占上级类别: %{customdata[1]:.2%}<extra></extra>"
    ),
    marker=dict(line=dict(width=1.5, color='rgba(255,255,255,0.8)'))
))

# 5. 布局与渲染
fig.update_layout(
    title=dict(text='消费支出层级占比旭日图', x=0.5, font=dict(size=16)),
    width=700,
    height=700,
    margin=dict(t=50, l=0, r=0, b=0),
    font=dict(family="Microsoft YaHei, SimHei, sans-serif")
)
fig.show()
# 兼容多环境渲染
# fig.show(renderer='browser') 
# 若在 Jupyter/VSCode 中仍不显示，可取消注释下方导出 HTML：
# fig.write_html("sunburst_final.html", include_plotlyjs=True)
```
