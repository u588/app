datasetsforecast 通常与另一个名为 statsforecast 的库协同工作。statsforecast 提供了模型定义、训练和预测的核心功能，而 datasetsforecast 则提供数据。

```python
# 1. 导入必要的库 
from datasetsforecast.m3 import M3 # 导入M3数据集 
from statsforecast import StatsForecast # 核心预测类 
from statsforecast.models  import AutoARIMA, TSB, AutoETS # 导入模型 
from statsforecast.utils  import ConformalIntervals # 用于预测区间 
```

```python
df, *_ = M3.load('data',  group='Monthly')
```

```python
df.tail(10)
```

```python
 
# 3. 定义你想要比较的模型列表 
models = [
    AutoARIMA(), # 自动ARIMA 
    AutoETS(),       # 指数平滑 
    # NHITS(h=12, input_size=24, # NHITS深度学习模型 
    #       loss='mae', max_steps=50)
]
```

```python
# 4. 初始化 StatsForecast 对象 
#    它负责处理数据和模型训练 
sf = StatsForecast(
    # df=df,
    models=models,
    freq='M', # 数据频率为 Monthly 
    n_jobs=-1 # 使用所有CPU核心并行计算 
)
```

```python
# 5. 进行预测 
#    h=12 表示预测未来12个时间点 
forecasts = sf.forecast(df=df,h=12,  level=[90]) # level 为置信区间 
```
