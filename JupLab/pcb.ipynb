```python
eng = create_engine('postgresql+psycopg://sa:11111111@10.3.18.56/tdxStocks')
```

```python
dfi = pd.read_sql('000001',engI)
```

```python
def pct(df,n,m):
    dff = df.copy()
    dff.set_index('datetime', inplace=True)
    dff['PCT21'] = dff['close'].pct_change(n)
    dff = dff[dff['PCT21']>m]
    return(dff)
```

```python
pct1 = pct(dfi,21,0.1)
fig = px.scatter(pct1,y='close',hover_name=pct1.index)
fig.update_layout(title=str(pct1.shape[0]))
fig.show()
```

```python
dfz = pd.read_sql('399001',engI)
pctz = pct(dfz,21,0.1)
fig = px.scatter(pctz,y='close',hover_name=pctz.index)
fig.update_layout(title=str(pctz.shape[0]))
fig.show()
```

```python
dfp = pd.read_sql('600409',eng)
pctp = pct(dfp,21,0.25)
fig = px.scatter(pctp,y='close',hover_name=pctp.index)
fig.update_layout(title=str(pctp.shape[0]))
# fig.update_xaxes(dtick='D',tickformat='%Y-%m-%d')
fig.show()
```

```python
def norm(df):
    min_value = df.min()
    max_value = df.max()
    nor = (df - min_value) / (max_value - min_value)
    return(nor)
```

```python
norm(pctz['close'])
```

```python
import plotly.graph_objects as go



# 创建散点图
fig = go.Figure()

# 添加第一个数据集
fig.add_trace(go.Scatter(x=pctz.index, y=norm(pctz['close']),hovertext=pctz['close'], mode='markers', name='数据集 1', marker=dict(color='blue')))

# 添加第二个数据集
fig.add_trace(go.Scatter(x=pctp.index, y=norm(pctp['close']),hovertext=pctp['close'], mode='markers', name='数据集 2', marker=dict(color='red')))

# 添加标题和轴标签
# fig.update_layout(
#     title="多列散点图",  # 图表标题
#     xaxis_title="X 轴标签",  # X 轴标签
#     yaxis_title="Y 轴标签"   # Y 轴标签
# )

# # 显示图表
fig.show()
```
