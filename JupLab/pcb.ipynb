```python
import pandas as pd
from sqlalchemy import create_engine
import plotly.express as px
```

```python
engI = create_engine('postgresql+psycopg://sa:11111111@10.3.18.56/tdxIndex')
```

```python
eng = create_engine('postgresql+psycopg://sa:11111111@10.3.18.56/tdxStocks')
```

```python
dfi = pd.read_sql('000001',engI)
```

```python
def pct(df,n,m):
    dff = df.copy()
    dff.set_index('datetime', inplace=True)
    dff['PCT21'] = dff['close'].pct_change(n)
    dff = dff[dff['PCT21']>m]
    return(dff)
```

```python
pct1 = pct(dfi,21,0.1)
```

```python
fig = px.scatter(pct1,y='close')
fig.show()
```
