```python
# scripts/verify_config_structure.py
"""
验证 V6.1 配置结构是否符合标准
"""
import sys
import os
from pathlib import Path
import pandas as pd
import numpy as np
from datetime import datetime

# 添加项目根目录到PATH
project_root = Path('../').resolve()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))
from infrastructure.base_services.config_service import ConfigService
```

```python
def verify_config_structure(config_path: str):
    """验证配置结构"""
    config = ConfigService(config_path)
    
    # 必需的根配置项
    required_root_configs = {
        '服务专属配置': [
            'commodity_thresholds',
            'commodity_strategy_map',
            'commodity_contracts',
            'composite_scoring',
            'alert_rules',
            'cross_market',
            'industry_rotation',
            'index_futures_contracts',
            'option_tolerance'
        ],
        'ThresholdService 配置': [
            'adaptive_config'
        ],
        '系统级配置': [
            'market_benchmarks',
            'risk_thresholds',
            'cache'
        ]
    }
    
    print("✅ V6.1 配置结构验证:")
    all_valid = True
    
    for category, keys in required_root_configs.items():
        print(f"\n{category}:")
        for key in keys:
            exists = key in config.config
            status = "✅" if exists else "❌"
            print(f"   {status} {key}: {'存在' if exists else '缺失'}")
            if not exists:
                all_valid = False
    
    if all_valid:
        print("\n🎉 所有必需配置项均在根配置层级！")
        print("✅ 可安全初始化所有服务（无配置缺失警告）")
    else:
        print("\n⚠️ 配置缺失，请补充上述缺失项")
    
    return all_valid
```

```python
verify_config_structure('system_config_v6.yaml')
```
