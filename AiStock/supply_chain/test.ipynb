```python
import networkx as nx
from pyvis.network import Network

# 1. 构建图
G = nx.karate_club_graph()
# 2. 初始化网络（notebook=False 生成独立 html）
net = Network(height="750px", width="100%", bgcolor="#1a1a2e", font_color="white", notebook=False)
net.from_nx(G)

# 3. 开启物理引擎（关键：调出 Neo4j 的弹簧拖拽感）
net.set_options("""
{
  "physics": {
    "enabled": true,
    "barnesHut": {"gravitationalConstant": -3000, "centralGravity": 0.3, "springLength": 90, "springConstant": 0.04},
    "minVelocity": 0.75,
    "solver": "barnesHut"
  },
  "interaction": {
    "hover": true,
    "tooltipDelay": 200,
    "dragNodes": true,
    "dragView": true,
    "zoomView": true
  }
}
""")

# 4. 导出
net.show(name="index.html", local=True, notebook=True)
```

```python
# 安装（如果尚未安装）
# !pip install pyvis

from pyvis.network import Network
import networkx as nx

# ============ 方式一：直接用 pyvis Network 构建 ============
net = Network(height="600px", width="100%", notebook=True, directed=False)

# 添加节点（id, label, color, size 等属性）
net.add_node(1, label="Alice",   color="#e74c3c", size=25, title="CEO")
net.add_node(2, label="Bob",     color="#3498db", size=20, title="CTO")
net.add_node(3, label="Carol",   color="#2ecc71", size=18, title="CFO")
net.add_node(4, label="Dave",    color="#f39c12", size=15, title="Engineer")
net.add_node(5, label="Eve",     color="#9b59b6", size=15, title="Designer")
net.add_node(6, label="Frank",   color="#1abc9c", size=12, title="Analyst")

# 添加边（source, target, value/宽度, label 等）
net.add_edge(1, 2, value=3, title="manages")
net.add_edge(1, 3, value=3, title="manages")
net.add_edge(2, 4, value=2, title="leads")
net.add_edge(2, 5, value=2, title="leads")
net.add_edge(3, 6, value=1, title="leads")
net.add_edge(4, 5, value=1, title="collaborates")
net.add_edge(4, 6, value=1, title="collaborates")

# 物理布局设置（让图更美观）
net.force_atlas_2based(gravity=-50, central_gravity=0.01, spring_length=150)

# 交互功能
net.show_buttons(filter_=["physics"])  # 显示物理参数调节面板

# 在 notebook 中直接展示
net.show("network_basic.html")

# # ============ 方式二：从 networkx 图转换 ============
# # 构建 networkx 图
# G = nx.karate_club_graph()

# # 转换为 pyvis 网络
# net2 = Network(height="600px", width="100%", notebook=True)
# net2.from_nx(G)

# # 按度数设置节点大小和颜色
# degrees = dict(G.degree())
# for node in net2.nodes:
#     d = degrees[node["id"]]
#     node["size"] = 10 + d * 3
#     node["title"] = f"Node {node['id']}\nDegree: {d}"
#     # 根据度数着色（度越大越红）
#     ratio = min(d / max(degrees.values()), 1.0)
#     r = int(255 * ratio)
#     b = int(255 * (1 - ratio))
#     node["color"] = f"rgb({r},80,{b})"

# net2.force_atlas_2based()
# net2.show("network_karate.html")
```
